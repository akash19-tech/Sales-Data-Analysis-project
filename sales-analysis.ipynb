# Add a city column

##### Convert Column to int

```python
#merged['City']=merged['Purchase Address'].apply(lambda x:x.split(',')[1])#this is for doing for one column but for the multiple column we can generte function for thatthing that we will do here
def city(address):
    return address.split(',')[1]
def state(address):
    return address.split(',')[2].split(' ')[1]# here we have split our data two times first on the ',' and second one on the blank space.

merged['City']=merged['Purchase Address'].apply(lambda x:city(x)+'('+state(x)+')')
#for doing this braket thing we can do one more thing that all the thing will be the same except after the x:  we can put  f"{city(x)}({state(x)})"
#that will do the same thing 
merged.head()
```

```python
dp=merged[merged['Order ID'].duplicated(keep=False)]
dp.head(20)
```

```python
dp['Grouped']=dp.groupby('Order ID')['Product'].transform(lambda x:','.join(x))#here we have made a new column by the name 'grouped' where we have used a groupby function where we have grouped on the basis of 'Order ID' column and do it with the column 'Product'
# transform we have used for applying some changes that we have to show in the new column we called as 'Ordered'
dp.head(20)
```

```python
dp=dp[['Order ID','Grouped']].drop_duplicates()
dp.head()
```

```python
from itertools import combinations
from collections import Counter

count = Counter()
for row in dp['Grouped']:
    row_list=row.split(',')
    count.update(Counter(combinations(row_list,2)))
    
count.most_common(10)
```
